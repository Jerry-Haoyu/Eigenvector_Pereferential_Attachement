```python
from epam.simulator import Simulator
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
N=50
A = np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-1), k=-1)

A_ = np.zeros((N+1, N+1))
A_[:N, :N] = np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-1), k=-1)
# A_[N, int(N/2)] = 1.0 
# A_[int(N/2), N] = 1.0


print(A_)
```

```python
model = Simulator(T=100, m=1, N=1, output_path="output/tmp")
model.simulate()
fig, ax = plt.subplots()
plot = ax.plot(np.arange(len(model.pfevec)), model.pfevec / np.sum(model.pfevec))[0]
ax.set_xlabel("Node Index")
ax.set_ylabel("Graph Probability Signal")
ax.set_title("Perturbation Of Probability Signal In a Line Graph")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
fig, ax = plt.subplots()
model = Simulator(T=10000, m=1, A_curr=A)
ax.set_xlabel("Node Index")
ax.set_ylabel("Graph Probability Signal")
ax.set_title("Perturbation Of Probability Signal In a Line Graph")
model.simulate()
ax.plot(np.arange(len(model.pfev)-1), model.pfev[:len(model.pfev)-1] / np.sum(model.pfev[:len(model.pfev)-1]))
plt.show()
```
